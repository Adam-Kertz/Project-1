```python
#Import pandas library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Define dataframe based on csv data (on BigFoot sightings)
bigfoot_geo_df = pd.read_csv("bfro_reports_geocoded.csv")
```

```python
#Create new dataframe copy and remove columns with large gaps in data
bigfoot_clean_df = bigfoot_geo_df.copy()
bigfoot_clean_df = bigfoot_clean_df.drop(columns=["uv_index","precip_type"])
```

```python
#Output clean data to csv file
bigfoot_clean_df.to_csv("Bigfoot Clean Data.csv")


#############################################################################################################################
# Mercedes' questions start in the cell below
##############################################################################################################################
```

```python
#Analyzing BigFoot sightings by 1st or 3rd hand

hand_group=bigfoot_clean_df.groupby("classification")
hand_group["classification"].count()

#Create the labels
labels=['First Hand','Second Hand','Third Hand']

#Capture sizes for each section
sizes=hand_group["classification"].count()

# The colors of each section of the pie chart
colors = ["lightskyblue", "lightgreen","white"]
```

```python
#Pie Chart
plt.pie(sizes, labels=labels, autopct="%1.1f%%", shadow=True, startangle=140,colors=colors)
plt.show()

#save as image
plt.savefig("../timothyrenner-bfro-sightings-data/hand_pie.png")
```
