```python
#Import pandas library
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import chi2_contingency
from scipy.stats import chisquare
```

#  BigFoot Sightings by Year

```python
#pull and read in the CSV file
Yearly_sightings_df = pd.read_csv("Bigfoot Clean Data.csv")
Yearly_sightings_df.head()
```

```python
Yearly_sightings_df.shape
```

```python
#show all columns
Yearly_sightings_df.columns
```

```python
# first BigFoot sighting date and state

first_sighting = Yearly_sightings_df.sort_values(by='date').iloc[0]
print("The first Bigfoot sighting was on", first_sighting['date'], "in", first_sighting['state'])
```

```python
# Convert the 'date' column to datetime format
Yearly_sightings_df['date'] = pd.to_datetime(Yearly_sightings_df['date'])

# Extract the year from the date column
Yearly_sightings_df['year'] = Yearly_sightings_df['date'].dt.year

# Group the sightings by year and count the number of sightings in each group
sightings_by_year = Yearly_sightings_df.groupby('year')['title'].count()

# Print the sightings by year
print(sightings_by_year)
```

```python
# organise sightings by year in descending order
sightings_by_year = sightings_by_year.sort_values(ascending=False)
print(sightings_by_year)
```

```python
print(sightings_by_year.describe())
```

```python
# Create a line plot of the sightings by year
plt.plot(sightings_by_year.index, sightings_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Sightings')
plt.title('BigFoot Sightings by Year')
plt.show()
```

```python
# Calculate total number of sightings
total_sightings = sightings_by_year.sum()

# Calculate proportion of sightings in each year
proportions = sightings_by_year / total_sightings

# Calculate expected values for each year
expected_values = total_sightings * proportions

# Print expected values
print(expected_values)
```

```python
# Calculate chi-square test statistic and p-value
chi2_stat, p_val = chisquare(sightings_by_year, f_exp=expected_values)

# Print results
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of freedom:", len(sightings_by_year) - 1)
print("Expected values:")
print(expected_values)
```

#### The Chi-square statistic of (1.081) is very small, which indicates that there is very little deviation between the observed and expected values.

#### The p-value is very high (1.0), which means that we cannot reject the null hypothesis that there is no association between the variables. Therefore, we can conclude that there is no significant association between the years and the number of sightings in each year.

#  Season (Time of Year) During Sightings

```python
# Extract the month from the date column
Yearly_sightings_df['month'] = Yearly_sightings_df['date'].dt.month

# Define the seasons
seasons = {1: 'Winter', 2: 'Winter', 3: 'Spring', 4: 'Spring', 5: 'Spring',
           6: 'Summer', 7: 'Summer', 8: 'Summer', 9: 'Fall', 10: 'Fall',
           11: 'Fall', 12: 'Winter'}

# Map the months to the seasons
Yearly_sightings_df['season'] = Yearly_sightings_df['month'].map(seasons)

# Group the sightings by season and count the number of sightings in each group
sightings_by_season = Yearly_sightings_df.groupby('season')['title'].count()

# show sightings_by_season
sightings_by_season
```

```python
# Plot the sightings by season
sightings_by_season.plot(kind='bar')

# Add labels and title
plt.xlabel('Season')
plt.ylabel('Number of sightings')
plt.title('BigFoot sightings by season')

# Add actual numbers on top of each bar
for i in range(len(sightings_by_season)):
    plt.text(i, sightings_by_season[i], sightings_by_season[i], ha='center')

plt.show()
```

```python
# Create a pie chart of the sightings by season
season_counts = Yearly_sightings_df['season'].value_counts()
plt.pie(season_counts, labels=season_counts.index, autopct='%1.1f%%')

# Add a title and legend to the chart
plt.title('Proportion of Sightings by Season')
plt.legend(title='Seasons')

# Show the chart
plt.show()
```

# Chi-square test of independence between the two categorical variables, "season" and "number of sightings."

```python
#The test compares the observed frequencies of sightings in each season to the expected frequencies assuming the null hypothese that there is no association between season and the number of sightings.

# Create a contingency table of the seasons and the number of sightings
cont_table = pd.crosstab(Yearly_sightings_df['season'], Yearly_sightings_df['title'])

# Get the observed values from the contingency table
observed_values = cont_table.values

# Calculate the chi-square statistic, p-value, degrees of freedom, and expected values
chi2, p, dof, expected_values = chi2_contingency(observed_values)

# Print the results
print(f"Chi-square statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected values:")
print(expected_values)
```

#### The Chi-square statistic is 11372.00, which is very large and indicates a significant difference between the observed and expected counts.

#### The P-value is 0.4877, which is greater than the typical significance level of 0.05. This means that we cannot reject the null hypothesis that there is no association between season and the number of sightings.
